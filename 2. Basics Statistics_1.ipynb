## Descriptive Analytics for Numerical Columns

To compute and analyze basic statistical measures for numerical columns in the dataset.

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
numerical_cols
```

```python
# Calculate basic statistics
descriptive_stats = pd.DataFrame({
    'Mean': df[numerical_cols].mean(),
    'Median': df[numerical_cols].median(),
    'Mode': df[numerical_cols].mode().iloc[0],
    'Standard Deviation': df[numerical_cols].std()
})
print(descriptive_stats)
```

## Steps that follow to statistical Measures for Numerical Columns
STEP 1: Load the dataset: Use pandas to read the data into a DataFrame.

STEP 2: Identify numerical columns: Check the data types to filter numerical columns.
Compute basic statistics.

STEP 3: Mean, median, mode, and standard deviation for numerical columns.
Use methods such as .mean(), .median(), .mode(), and .std().

# Data Visualization

## HISTOGRAMS

```python
# Plot histograms
for col in numerical_cols:
    plt.hist(df[col], bins=20, edgecolor='k')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col,color='red',size=15)
    plt.ylabel('Frequency',color='green',size=15)
    plt.show()
```

## BOXPLOTS

```python
# Plot boxplots
for col in numerical_cols:
    plt.boxplot(df[col].dropna(), vert=False)
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

## BAR CHARTS FOR CATEGORICAL VARIABLES

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Plot bar charts for categorical columns
for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Category Distribution: {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## STANDARDIZATION OF NUMERICAL VARIABLES

```python
# Standardize numerical columns
standardized_df = (df[numerical_cols] - df[numerical_cols].mean()) / df[numerical_cols].std()

# Before and after comparison
print("Before Standardization:")
print(df[numerical_cols].describe())
print("\nAfter Standardization:")
print(standardized_df.describe())
```

## CONVERSION OF CATEGORICAL DATA INTO DUMMY VARIABLES

```python
# Apply one-hot encoding
encoded_df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Display a portion of the transformed dataset
print(encoded_df.head())
```
